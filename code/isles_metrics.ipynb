```python
import nibabel as nib
import numpy as np
import os
import eval_utils
```

```python
gt_path = "/media/cansu/DiskSpace/Cansu/ISLES24/ISLES24-Multimodal/data/challenge_test_set/test/results/sub-stroke0018_ses-01_all_modalities_lab.nii.gz"
pred_path = "/media/cansu/DiskSpace/Cansu/ISLES24/ISLES24-Multimodal/data/challenge_test_set/test/results/sub-stroke0018_ses-01_all_modalities_pred.nii.gz"

results_path = "/media/cansu/DiskSpace/Cansu/ISLES24/ISLES24-Multimodal/data/challenge_test_set/test/results"

# create a csv file to store the results

for file in os.listdir(results_path):
    if file.endswith("_lab.nii.gz"):
        gt_path = os.path.join(results_path, file)
        pred_path = gt_path.replace("_lab.nii.gz", "_pred.nii.gz")
        gt_image = nib.load(gt_path).get_fdata()
        segmented_image = nib.load(pred_path).get_fdata()

        # Compute F1-score, lesion count difference and Dice/
        f1, lcd, dice = eval_utils.compute_dice_f1_instance_difference(gt_image, segmented_image)
        print('Absolute lesion count difference: {} '.format(lcd))
        print('Lesion-wise F1-score: {} '.format(f1))
        print('Dice-score: {} '.format(dice))

        # Compute absolute volume difference
        voxel_volume = np.prod(nib.load(gt_path).header.get_zooms())/1000 # Get voxel volume
        print('Absolute volume difference: {} ml'.format(eval_utils.compute_absolute_volume_difference(gt_image, segmented_image, voxel_volume)))

        
```
